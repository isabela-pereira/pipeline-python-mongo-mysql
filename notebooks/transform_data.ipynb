## Aplicando transformações: filtrando registros da categoria livros

Agora, vamos executar uma consulta para obter apenas os dados relacionados à categoria de **livros**.

```python
query = { "Categoria": "electronics" }

for doc in collection.find(query):
    print(doc)
```

```python
query = { "Categoria": "electronics" }

lista_eletronicos = []
for doc in collection.find(query):
    #print(doc)
    lista_eletronicos.append(doc)
```

```python
lista_eletronicos
```

**Salvando os dados em um DataFrame**

```python
import pandas as pd

df_eletronicos = pd.DataFrame(lista_eletronicos)
df_eletronicos.head()
```

```python

df_rating = df_eletronicos['Avaliação'].apply(pd.Series)

# 3. Renomeie as colunas dentro desse DataFrame temporário
df_rating = df_rating.rename(columns={'rate': 'Nota', 'count': 'Total_Avaliacoes'})

# 4. Junte (concatene) as novas colunas ao DataFrame original
# Removemos a coluna 'Avaliação' antiga para não ficar duplicado
df_final = pd.concat([df_eletronicos.drop('Avaliação', axis=1), df_rating], axis=1)

# Visualize o resultado final
df_final.head()
```

## Aplicando transformações: filtrando produtos vendidos a partir de 2021

Nesta etapa, faremos uma query para selecionar apenas os produtos vendidos a partir do ano de 2021. Utilizaremos um filtro na coluna de data de compra para obter os registros que atendam a essa condição.

```python
collection.find_one()
```

```python
# Corrigindo o nome do campo para 'Avaliação' e acessando a subchave 'rate'
query = {"Avaliação.rate": {"$gte": 4.0, "$lt": 5.0}} 
lista_produtos = []
for doc in collection.find(query):
    lista_produtos.append(doc)
```

```python
lista_produtos
```

```python
df_produtos = pd.DataFrame(lista_produtos)
df_produtos.head()
```

**Formatando as datas**

```python

df_rating = df_produtos['Avaliação'].apply(pd.Series)

# 3. Renomeie as colunas dentro desse DataFrame temporário
df_rating = df_rating.rename(columns={'rate': 'Nota', 'count': 'Total_Avaliacoes'})

# 4. Junte (concatene) as novas colunas ao DataFrame original
# Removemos a coluna 'Avaliação' antiga para não ficar duplicado
df_final = pd.concat([df_eletronicos.drop('Avaliação', axis=1), df_rating], axis=1)

# Visualize o resultado final
df_final.head()
```

```python
import os

# 1. Define o caminho da pasta
diretorio = '../avaliacao'

# 2. Verifica se a pasta existe; se não, cria ela
if not os.path.exists(diretorio):
    os.makedirs(diretorio)
    print(f"Pasta '{diretorio}' criada com sucesso!")

# 3. Agora você pode salvar sem erro
df_final.to_csv(f'{diretorio}/tb_4_5_em_diante.csv', index=False)
```
